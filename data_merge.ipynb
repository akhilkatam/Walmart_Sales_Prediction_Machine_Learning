```python
import numpy as np
import pandas as pd
```

```python
sal_df = pd.read_csv('sales_train_validation.csv')
cal_df = pd.read_csv('calendar.csv')
pri_df = pd.read_csv('sell_prices.csv')
```

```python
cal_df.head()
```

```python
sal_df.head()
```

```python
pri_df.head()
```

```python
melted = pd.melt(sal_df, var_name='d', value_name='sales', id_vars=['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'])
melted.head(20)
```

```python
melted.shape
```

```python
melted['id'] = melted['id'].astype('category')
melted['item_id'] = melted['item_id'].astype('category')
melted['dept_id'] = melted['dept_id'].astype('category')
melted['cat_id'] = melted['cat_id'].astype('category')
melted['store_id'] = melted['store_id'].astype('category')
melted['state_id'] = melted['state_id'].astype('category')

melted.head(20)
```

```python
merged = melted.merge(cal_df, on='d', how='left')
merged.head(20)
```

```python
merged = merged.merge(pri_df, on=['store_id', 'item_id', 'wm_yr_wk'], how='left')
merged.head(20)
```

```python
merged.shape
```

```python
merged.to_csv("merged_data.csv", encoding='utf-8', index=False)
```
