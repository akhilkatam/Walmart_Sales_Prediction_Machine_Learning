```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for column in dataframe.columns:
    dtype = dataframe[column].dtypes
    if dtype == 'category':
        print("Transforming", column)
        le.fit(dataframe[column])
        dataframe[column] = le.transform(dataframe[column])
dataframe.head()
```

```python
#dropping date for now cuz error
dataframe = dataframe.drop(labels='date', axis=1)
```

```python
#split train and test
train = dataframe[dataframe['d']<1885]
test = dataframe[dataframe['d'] >= 1885]
```

```python
#splitting train data and train label
train_label = train['sales']
train = train.drop(labels='sales', axis=1)

#splitting test data and test label
test_label = test['sales']
test = test.drop(labels='sales', axis=1)
```

```python
test_label = test_label.values
```

```python
from xgboost import XGBRegressor
import datetime

print(datetime.datetime.now())

model = XGBRegressor(verbosity=2, max_depth=7)
model.fit(train,train_label)

print(datetime.datetime.now())
```

```python
predictions = model.predict(test)
```

```python
for i in range(len(test_label)):
    print(predictions[i], test_label[i])
```

```python
pdp = pd.DataFrame(predictions)
pdp.to_csv("6years_xg_depth7_predictions.csv")
```
