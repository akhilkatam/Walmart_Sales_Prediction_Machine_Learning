```python
import pandas as pd
import numpy as np
import tqdm as tqdm
```

```python
dataframe = pd.read_pickle('reduced.pickle')
#pre processing missing values of prices by transforming with mean price of that id
dataframe['sell_price'].fillna(dataframe.groupby('id')['sell_price'].transform('mean'),inplace=True)
dataframe.head(100)
```

```python
#insert lag features to 'dataframe'
shifting = 28
for i in range(9): #num of weeks to shift here 8 weeks we consider
    dataframe['lag_'+str(shifting+(7*i))] = dataframe.groupby('id')['sales'].shift(shifting+(7*i)).astype(np.float16)
```

```python
dataframe.tail(100)
```

```python
#insert rolling features 'dataframe'
for i in [7,14,28,35,60]:
    dataframe['rolling_mean_'+str(i)] =  dataframe.groupby(['id'])['lag_28'].transform(lambda x: x.rolling(i).mean())
    dataframe['rolling_median_'+str(i)] =  dataframe.groupby(['id'])['lag_28'].transform(lambda x: x.rolling(i).median()) 
```

```python
dataframe.tail(1000)
```

```python
#reduce size of lag/rolling features
dataframe.info(memory_usage='deep')
```

```python
for col in dataframe.columns[-10:]:
    print(dataframe[col].max())
    
for col in dataframe.columns[-10:]:
    print(dataframe[col].min())
```

```python
dataframe['rolling_mean_7'] = dataframe['rolling_mean_7'].astype('float16')
dataframe['rolling_mean_14'] = dataframe['rolling_mean_14'].astype('float16')
dataframe['rolling_mean_28'] = dataframe['rolling_mean_28'].astype('float16')
dataframe['rolling_mean_35'] = dataframe['rolling_mean_35'].astype('float16')
dataframe['rolling_mean_60'] = dataframe['rolling_mean_60'].astype('float16')
dataframe['rolling_median_7'] = dataframe['rolling_median_7'].astype('float16')
dataframe['rolling_median_14'] = dataframe['rolling_median_14'].astype('float16')
dataframe['rolling_median_28'] = dataframe['rolling_median_28'].astype('float16')
dataframe['rolling_median_35'] = dataframe['rolling_median_35'].astype('float16')
dataframe['rolling_median_60'] = dataframe['rolling_median_60'].astype('float16')
```

```python
for col in dataframe.columns[-10:]:
    print(dataframe[col].max())
    
for col in dataframe.columns[-10:]:
    print(dataframe[col].min())
```
